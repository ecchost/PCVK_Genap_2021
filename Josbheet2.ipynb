```
class MyImage:
    def __init__(self, img_name):
        self.img = cv.imread(img_name)
        self.__name = img_name.split('/')[5]

    def name(self):
        return self.__name

src = MyImage('/content/drive/MyDrive/PCVK/kucing2balik.jpeg')
# extract blue channel
blue_channel = src.img[:,:,0]

# create empty image with same shape as that of src image
blue_img = np.zeros(src.img.shape)

#assign the red channel of src to empty image
blue_img[:,:,0] = blue_channel

cv.rectangle(src.img, pt1=(10,350), pt2=(100,250), color=(0,0,250), thickness = 5)

font = cv.FONT_ITALIC
cv.putText(src.img, text=src.name(), org=(400,300), fontFace=font, fontScale=1, color=(255,0,0), thickness=6)
plt.imshow(src.img)
```
